Step 1:

Use code from the paper (provided in Assignment 07) to clean the NHANES dietary data and cluster food items into food groups using the same parameters reported in the paper

```python
import os

# Create a Spark Context
if os.name == 'nt':
    import findspark
    findspark.init(r"D:\Downloads\spark-2.2.1-bin-hadoop2.7")

from pyspark import SparkContext
sc = SparkContext.getOrCreate()

# Load code from paper
from src.preprocess import cleanNHANESData
from src.DBSCAN import DBSCAN

# Load NHANES dietary data from 2009 and 2011
# The preprocessing code expects a dictionary with RDDs of NHANES dietary data
raw_data = {}
for year in ['1999_DRXIFF', '2001_DRXIFF', '2003_DR1IFF', '2003_DR2IFF', '2005_DR1IFF', '2005_DR2IFF', '2007_DR1IFF', '2007_DR2IFF', '2009_DR1IFF', '2009_DR2IFF', '2011_DR1IFF', '2011_DR2IFF']:
    raw_data[year] = sc.textFile('./NHANES_dietary_data/{}.csv'.format(year))
    
# Print total number of food item entries
total = 0
for year_data in raw_data.values():
    total += year_data.count()
print('Total number of food entries: {}'.format(total))
    
# Load list of nutrient features we want from the raw dietary data
with open('./NHANES_dietary_data/features.txt') as f:
    features = f.readlines()
features = [f.strip() for f in features]

# Look at the list of features you will use for each food item
# We need this list because some features (i.e., values) in the raw NHANES data cannot be used to cluster
# Note that the first 4 features are:
# "USDA food code", "Respondent sequence number", "Modification code", and "Grams"
# The remaining features are micro- and macronutrient features
print('List of features:\n', features)
```

```python
# Apply preprocessing code from paper
clean_data = cleanNHANESData(sc, raw_data, features).persist()

# Look at cleaned NHANES dietary data
# Elements in RDD are key-value pairs of (food ID, np.array(nutrient value))
print ('Total unique food items: {}'.format(clean_data.count()))
print(clean_data.take(1))
```

```python
# Define DBSCAN parameters
epsilon = 1
min_pts = 4
metric = 'euclidean'

# Cluster the food items with DBSCAN
food_clusters = DBSCAN(sc, clean_data, epsilon=epsilon, minpts=min_pts, metric=metric)

cluster_data = food_clusters.join(clean_data)
#print(cluster_data.take(1))
reduced_data= cluster_data.map(lambda x: (x[0], x[1][1]) )


# Isolate cluster ID and nutrient value (i.e., get rid of food item ID)
# RDD should be <<(cluster ID, [nutrient, values]), (cluster ID, [nutrient, values]), ...>>
cluster_nutrient = cluster_data.values()
#print(cluster_nutrient.countByKey())
# Look at the cluster results
# Elements in the RDD are key-value pairs of (food ID, cluster ID)
print('Clustered food items: {}'.format(food_clusters.count()))
print(food_clusters.take(1))
```

```python
# Define a function to report the percent of total food items that are clustered
# The input 'clusters' should be the output of DBSCAN
# The input 'total' should be the total number of unique food items (before clustering)
def percentClustered(clusters, total):
    pc= (clusters.count() / clean_data.count())*100
    return pc
    
# Define a function to count the number of clusters
# The input 'clusters' should be the output of DBSCAN
def clusterCount(clusters):
    cc= clusters.map(lambda x: (x[1], x[0]))
    cc= cc.reduceByKey(lambda x,y: x).count()
    return cc
```

```python
# Apply the functions you defined to the clusters found above
perc_clust = percentClustered(food_clusters, clean_data.count())
clust_cnt = clusterCount(food_clusters)

# Print results
print('Food items clustered: {:.2f}%'.format(perc_clust))
print('Number of food clusters: {}'.format(clust_cnt))
```

```python
# Define DBSCAN parameters
epsilon = 0.08
min_pts = 4
metric = 'cosine'

# Cluster the food items with DBSCAN
food_clusters2 = DBSCAN(sc, clean_data, epsilon=epsilon, minpts=min_pts, metric=metric)
cluster_data2 = food_clusters2.join(clean_data)

# Isolate cluster ID and nutrient value (i.e., get rid of food item ID)
# RDD should be <<(cluster ID, [nutrient, values]), (cluster ID, [nutrient, values]), ...>>
cluster_nutrient2 = cluster_data2.values()

# Look at the cluster results
# Elements in the RDD are key-value pairs of (food ID, cluster ID)
print('Clustered food items: {}'.format(food_clusters2.count()))
print(food_clusters2.take(1))
```

```python
# Apply the functions you defined to the clusters found above
perc_clust = percentClustered(food_clusters2, clean_data.count())
clust_cnt = clusterCount(food_clusters2)

# Print results
print('Food items clustered: {:.2f}%'.format(perc_clust))
print('Number of food clusters: {}'.format(clust_cnt))
```

Step 3:

Write Spark code to measure inter- and intra-cluster distance of food groups

```python
import sys
from src.search import PPDistance

# The inter-cluster distance is the minimum distance between all foods in X and all foods outside X
def InterClusterDist(x, clusters, metric):
    clusters = clusters.value
    key = x[0]
    value = x[1]
    minimum_distance = sys.maxsize

    for point in clusters:
        if point[0] == key:
            continue

        distance = PPDistance(value, point[1], metric=metric)
        
        if distance < minimum_distance:
            minimum_distance = distance
    
    return minimum_distance

# The intra-cluster distance is the maximum distance between any two foods in X
def IntraClusterDist(x, clusters, metric):
    clusters = clusters.value
    key = x[0]
    value = x[1]
    maximum_distance = 0
    
    for point in clusters:
        if point[0] != key:
            continue
            
        distance = PPDistance(value, point[1], metric=metric)
        
        if distance > maximum_distance:
            maximum_distance = distance
            
    return maximum_distance
```

Step 4:

Measure the inter- and intra-cluster distances for each food group in the food groups found with DBSCAN, 2-digit USDA, and 3-digit USDA.  Plot these values in boxplots to replicate Figures 10a and 10b from the original paper


```python
# Define function for measuring distances
import numpy as np

def measureDistances(data, metric):
    # Broadcast the clusters
    cluster_broadcast = data.map(lambda x: x[1])
    cluster_broadcast = cluster_broadcast.collect()
    cluster_broadcast = sc.broadcast(cluster_broadcast)
    
    # make the data in correct format
    data = data.map(lambda x: (x[1][0], x[1]))
    
    # intracluster distances
    intra = data.mapValues(lambda x: IntraClusterDist(x, cluster_broadcast, metric))
    intra = intra.reduceByKey(lambda x,y: max(x,y))
    intra = intra.map(lambda x: x[1])
    intra = intra.collect()   
    
    # intercluster distances
    inter = data.mapValues(lambda x: InterClusterDist(x, cluster_broadcast, metric))
    inter = inter.reduceByKey(lambda x,y: min(x,y))
    inter = inter.map(lambda x: x[1])
    inter = inter.collect()
    
    return intra,inter,metric


DBSCAN_intra_e,DBSCAN_inter_e,DBSCAN_e = measureDistances(cluster_data, 'euclidean')
DBSCAN_intra_c,DBSCAN_inter_c,DBSCAN_c = measureDistances(cluster_data, 'cosine')
D2_intra_e,D2_inter_e,D2_e = measureDistances(D2_cluster, 'euclidean')
D2_intra_c,D2_inter_c,D2_c = measureDistances(D2_cluster, 'cosine')
D3_intra_e,D3_inter_e,D3_e = measureDistances(D3_cluster, 'euclidean')
D3_intra_c,D3_inter_c,D3_c = measureDistances(D3_cluster, 'cosine')
```

```python
import matplotlib.pyplot as plt
# function to plot the data ( boxplot )
def plotData(d1,d2,d3,m,d):
    data = [d1,d2,d3]
    plt.boxplot(data,0,'',whis= [10,90],showmeans=True)
    xvals=['DBSCAN','2-digit','3-digit']
    plt.xticks(1+np.arange(len(xvals)), xvals)
    if d=='inter':
        plt.xlabel('Cluster Separation {} | (Higher Better)'.format(m))
        plt.ylabel('Separation (min)')
    else:
        plt.xlabel('Cluster Diameters {} | (Lower Better)'.format(m))
        plt.ylabel('Diameter')
    plt.show()
```

```python
# Cluster Separation Comparison using Euclidean
plotData(DBSCAN_inter_e,D2_inter_e,D3_inter_e,DBSCAN_e,'inter')
print("Mean:")
print(np.mean(DBSCAN_inter_e))
print(np.mean(D2_inter_e))
print(np.mean(D3_inter_e))
print("Median")
print(np.median(DBSCAN_inter_e))
print(np.median(D2_inter_e))
print(np.median(D3_inter_e))

```

```python
# Cluster Separation Comparison using Cosine
plotData(DBSCAN_inter_c,D2_inter_c,D3_inter_c,DBSCAN_c,'inter')
print("Mean:")
print(np.mean(DBSCAN_inter_c))
print(np.mean(D2_inter_c))
print(np.mean(D3_inter_c))
print("Median")
print(np.median(DBSCAN_inter_c))
print(np.median(D2_inter_c))
print(np.median(D3_inter_c))
```

```python
# Cluster Diameter Comparison using Euclidean
plotData(DBSCAN_intra_e,D2_intra_e,D3_intra_e,DBSCAN_e,'intra')
print("Mean:")
print(np.mean(DBSCAN_intra_e))
print(np.mean(D2_intra_e))
print(np.mean(D3_intra_e))
print("Median:")
print(np.median(DBSCAN_intra_e))
print(np.median(D2_intra_e))
print(np.median(D3_intra_e))
```

```python
from sklearn.manifold import TSNE


#project the data (and clusters) down to 2 dimensions and visualize the clusters
def plotFoodClusters2D(clusters):
    # Get cluster IDS to color plotted clusters
    cluster_ids = clusters.keys().collect()
    unique_ids = np.unique(cluster_ids).tolist()
    cluster_ids = [unique_ids.index(i) for i in cluster_ids]
    
    # Reduce nutrient values to 2 dimensions with TSNE
    # HINT: ref [1,2]
    nutrient_values = clusters.values().collect()
    cluster_embedded = TSNE(n_components=2).fit_transform(nutrient_values)
    
    # Plot clusters
    X = cluster_embedded[:,0]
    Y = cluster_embedded[:,1]
    plt.scatter(X, Y, c=cluster_ids, s=3)
    plt.xticks(np.arange(-80, 100, step=20))
    plt.yticks(np.arange(-100, 100, step=20))

    plt.show()
    plt.clf()
    

plotFoodClusters2D(cluster_nutrient)
plotFoodClusters2D(D2_nutrient)
plotFoodClusters2D(D3_nutrient)



```
